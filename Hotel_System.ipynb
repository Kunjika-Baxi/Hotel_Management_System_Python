```python
import mysql.connector as m
db=m.connect(host="localhost",user="root",passwd="root")
cur=db.cursor()
cur.execute("create database Taj_db")
cur.execute("use Taj_db")
cur.execute("create table if not exists rooms (room_no int primary key, room_type varchar(30),capacity int not null,charges float not null, room_status varchar(30))")
cur.execute("create table if not exists booking (booking_id int auto_increment primary key,name varchar(40) not null,contact varchar(11) not null,email varchar(25),address varchar(70),no_of_person int not null, room_no int references rooms(room_no),check_in date not null, check_out date not null,bill_amount float)")
cur.execute("alter table rooms add constraint check(room_status in ('available','occupied'))")
db.commit()
```
